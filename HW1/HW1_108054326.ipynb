```python
import math
number = int(input('Input a positive integer N(N < 2^31 - 1)'))
ans = 1
#search prime number by dividing from one to the sqrt of number
for i in range(2,int(math.sqrt(number))+1):
    if number % i == 0:
        ans = 0
if ans == 1:
    print('YES')
else:
    print('NO')
```

```python
value = int(input('Input a positive integer N(N < 2^31 - 1)'))
total = []
power = len(str(value))
digits = [int(x) for x in str(value)]
for i in digits:
    product = i**power
    total.append(product)
    sum_total = sum(total)
if(sum_total == value):
    print('YES')
else:
    print('NO')
```

```python
number = int(input('Input a positive integer N ( N < 100000)'))
List = [int(number) for number in input().split()]
answer = 0
for i in (0,len(List)):
    if List.count(List[i-1]) > answer:
        answer = List.count(List[i])
if answer == 1:
    print('0')
else:
    print(answer)
```
